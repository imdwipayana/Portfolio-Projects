```python
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler

import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_labels = []
data_age = []
data_salary = []
```

```python
for i in range(50):
    # The ~5% of younger individuals who do not have insurance
    random_younger = randint(18,64)
    young_salary_random = randint(10,49)
    data_age.append(random_younger)
    data_salary.append(young_salary_random*100000)
    data_labels.append(0)
    
    # The ~5% of older individuals who has insurance
    random_older = randint(65,100)
    old_salary_random = randint(50,100)
    data_age.append(random_older)
    data_salary.append(old_salary_random*100000)
    data_labels.append(1)
    
for i in range(1000):
    # The ~95% of younger individuals who has insurance
    random_younger = randint(18,64)
    young_salary_random = randint(50,300)
    data_age.append(random_younger)
    data_salary.append(young_salary_random*100000)
    data_labels.append(1)
    
    # The ~95% of older individuals who did not has insurance
    random_older = randint(65,100)
    old_salary_random = randint(0,49)
    data_age.append(random_older)
    data_salary.append(old_salary_random*100000)
    data_labels.append(0)
```

```python
data_age = np.array(data_age)
data_salary = np.array(data_salary)
data_labels = np.array(data_labels)
data_age, data_salary, data_labels = shuffle(data_age,data_salary,data_labels)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data_age = scaler.fit_transform(data_age.reshape(-1,1))
scaled_data_salary = scaler.fit_transform(data_salary.reshape(-1,1))
```

```python
data_train = pd.DataFrame({'Age':data_age, 'Salary':data_salary, 'Insurance':data_labels})
```

```python
data_train.head()
```

```python
data_train_pisah = data_train[['Age','Salary']]
```

```python
data_target = data_train[['Insurance']]
```

```python
data_train_pisah, data_target = shuffle(data_train_pisah, data_target)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data_age = scaler.fit_transform(data_train_pisah)
```

```python
scaled_data_age
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaled_data_age,data_target,test_size=0.2,random_state=5)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

```python
model = Sequential([
    Dense(units=16, input_shape=(2,), activation='relu'),
    Dense(units=32, activation='relu'),
    Dense(units=2, activation='softmax')
])
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001),loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x=X_train, y=y_train, validation_split=0.1,batch_size=10,epochs=30,shuffle=True,verbose=2)
```

```python
predictions = model.predict(x=X_test, batch_size=10, verbose=0)
```

```python
model.evaluate(X_test,y_test)
```

```python
rounded_predictions = np.argmax(predictions, axis=-1)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test,rounded_predictions))
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=y_test,predictions=rounded_predictions)

plt.figure(figsize = (10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
